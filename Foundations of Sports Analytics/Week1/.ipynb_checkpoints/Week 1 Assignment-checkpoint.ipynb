```python
# Load the packages
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data. 
# EPL results for the 2017/18 season

EPL18 = pd.read_excel('Assignment Data/Week 1/EPL2017-18.xlsx')
print(EPL18.columns.tolist())
```

```python
# Visualizing Premier League (2017/2018) dataset
EPL18.head()
```

## Step 2

```python
EPL18['hwinvalue']=np.where(EPL18['FTR']=='H',1,np.where(EPL18['FTR']=='D',.5,0))
EPL18['awinvalue']=np.where(EPL18['FTR']=='A',1,np.where(EPL18['FTR']=='D',.5,0))
EPL18['count']=1
```

```python
EPL18.head()
```

## Step 3

```python
EPL_half1 = EPL18[EPL18.Date < 20180000]
EPL_half1.head()
```

```python
EPL_half2 = EPL18[EPL18.Date >= 20180000]
EPL_half2.head()
```

```python
Enghome1 = EPL_half1.groupby(['HomeTeam'])['count','hwinvalue', 'FTHG','FTAG'].sum().reset_index()
Enghome1 = Enghome1.rename(columns={'HomeTeam':'team','count':'Ph','FTHG':'FTHGh','FTAG':'FTAGh'})
Enghome1.head()
```

```python
Engaway1 = EPL_half1.groupby(['AwayTeam'])['count','awinvalue', 'FTHG','FTAG'].sum().reset_index()
Engaway1 = Engaway1.rename(columns={'AwayTeam':'team','count':'Pa','FTHG':'FTHGa','FTAG':'FTAGa'})
Engaway1
```

## Step 4 (home team)

```python
# Merge the home team and away team results

Eng17 = pd.merge(Enghome1, Engaway1, on = ['team'])
Eng17
```

## Step 5 (home team)

```python
Eng17['W'] = Eng17['hwinvalue']+Eng17['awinvalue']
Eng17['G'] = Eng17['Ph']+Eng17['Pa']
Eng17['GF'] = Eng17['FTHGh']+Eng17['FTAGa']
Eng17['GA'] = Eng17['FTAGh']+Eng17['FTHGa']
Eng17
```

```python
# Order teams by FTHGh

sns.catplot(y='team', x='FTHGh', kind='bar',order=Eng17.sort_values('FTHGh').team, palette="ch:start=.5, rot=-.75", data=Eng17)
plt.title('Ordering by team FTHGh', size=14)


sns.catplot(y='team', x='FTHGa', kind='bar',order=Eng17.sort_values('FTHGa').team, palette="ch:start=.6, rot=-.46", data=Eng17)
plt.title('Ordering by team FTHGa', size=14)
```

```python
# Order teams by GF in first half of season
sns.catplot(y='team', x='GF', kind='bar',order=Eng17.sort_values('GF', ascending=False).team, palette="ch:start=.4, rot=-.30", data=Eng17)
plt.title('Ordering by team GF in home', size=14)
```

```python
# Order teams by GA in first half of season

sns.catplot(y='team', x='GA', kind='bar',order=Eng17.sort_values('GA', ascending=False).team, palette="ch:start=.7, rot=-.22", data=Eng17)
plt.title('Ordering by team GA in home', size=14)
```

## Step 6 (home team)

```python
# Create the win percentage and Pythagorean Expectation

Eng17['wpc'] = Eng17['W']/Eng17['G'] 
Eng17['pyth'] = Eng17['GF']**2/(Eng17['GF']**2 + Eng17['GA']**2) 
Eng17.head()
```

```python
Eng17 = Eng17.rename(columns={'wpc':'wpc_2017','pyth':'pyth_2017'})
Eng17.head()
```

## Step 7 (=Step 4 (away team))

```python
Enghome2 = EPL_half2.groupby(['HomeTeam'])['count','hwinvalue', 'FTHG','FTAG'].sum().reset_index()
Enghome2 = Enghome2.rename(columns={'HomeTeam':'team','count':'Ph','FTHG':'FTHGh','FTAG':'FTAGh'})
Enghome2.head()
```

```python
Engaway2 = EPL_half2.groupby(['AwayTeam'])['count','awinvalue', 'FTHG','FTAG'].sum().reset_index()
Engaway2 = Engaway2.rename(columns={'AwayTeam':'team','count':'Pa','FTHG':'FTHGa','FTAG':'FTAGa'})
Engaway2.head()
```

```python
# Merge the home team and away team results

Eng18 = pd.merge(Enghome2, Engaway2, on = ['team'])
Eng18.head()
```

## Step 7 (=Step 5 (away team))

```python
Eng18['W'] = Eng18['hwinvalue']+Eng18['awinvalue']
Eng18['G'] = Eng18['Ph']+Eng18['Pa']
Eng18['GF'] = Eng18['FTHGh']+Eng18['FTAGa']
Eng18['GA'] = Eng18['FTAGh']+Eng18['FTHGa']
Eng18
```

## Step 7 (=Step 6 (away team))

```python
# Create the win percentage and Pythagorean Expectation

Eng18['wpc'] = Eng18['W']/Eng18['G'] 
Eng18['pyth'] = Eng18['GF']**2/(Eng18['GF']**2 + Eng18['GA']**2) 
Eng18.head()
```

```python
Eng17.head()
```

```python
Eng18 = Eng18.rename(columns={'wpc':'wpc_2018','pyth':'pyth_2018'})
Eng18
```

```python
sns.catplot(y='team', x='W', kind='bar',order=Eng18.sort_values('W', ascending=False).team, palette="ch:start=.7, rot=-.22", data=Eng18)
```

## Step 8

```python
# Now merge 2017 and 2018 summary files.
EPL1718 = pd.merge(Eng17, Eng18, on = ['team'])
EPL1718.head()
```

## Step 9

```python
# First, plot Pythagorean Expectation against win percentage in the second half of the season

sns.relplot(x="pyth_2017", y="wpc_2018", data = EPL1718)
```

```python
# Now, compare this with a plot of win percentage from the first half of the season against win percentage
#in the second half of the season


sns.relplot(x="wpc_2017", y="wpc_2018", data = EPL1718)
```

```python
# The two plots look similar
# We can be more precise still if we compare the correlation coefficients. The first row of the table shows the 
# correlation of win percentage in second half of the season against itself, win percentage in the first half of the season,
# Pythagorean Expectation in the first half of the season, and Pythagorean Expectation in the second half of the season.
# Our focus is on comparing the second and third columns.

keyvars = EPL1718[['team','wpc_2017','wpc_2018','pyth_2017','pyth_2018']]
keyvars.corr()
```

```python
corr = keyvars.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
